# Calculating the prediction

```python
import pandas as pd
import pickle
```

## Input new data

```python
test_file = 'list_to_predict.txt'
```

```python
load_data = pd.read_table(test_file, sep=' ', header=None)
```

```python
desc = pd.read_csv('fingerprint_output_file.csv')
desc
```

```python
desc.shape
```

```python
Xlist = list(pd.read_csv('descriptor_list.csv').columns)
desc_subset = desc[Xlist]
desc_subset
```

```python
desc_subset.shape
```

```python
load_model = pickle.load(open('pickle_model.pkl', 'rb'))
prediction = load_model.predict(desc_subset)
prediction_output = pd.Series(prediction, name='pIC50')
molecule_name = pd.Series(load_data[1], name='molecule_name')
df_pred = pd.concat([molecule_name, prediction_output], axis=1)
df_pred
```

```python
load_data.columns = ["canonical_smiles","molecule_chembl_id"]
load_data_cols = load_data
df_pred_comb = pd.concat([load_data_cols,prediction_output], axis=1)
```

```python
import mols2grid

mols2grid.display(df_pred_comb,
                  # RDKit's MolDrawOptions parameters
                  fixedBondLength=25,
                  # rename fields for the output document
                  rename={"canonical_smiles": "SMILES", "molecule_chembl_id": "Name"},
                  # set what's displayed on the grid
                  subset=["Name","img","pIC50"],
                  # set what's displayed on the tooltips
                  tooltip=["Name","pIC50"],
                  # style for the grid labels and tooltips
                  #style={"pIC50": lambda x: "color: red" if x < -3 else "color: black"},
                  # change the precision and format (or other transformations)
                  transform={"pIC50": lambda x: f"{x:+.2f}"})
```
